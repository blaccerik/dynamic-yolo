Notebook to use VOC data and make it a bit smaller

```python
import os

data_output = "data"
data_input = "datasets/VOC"

dirs = [
    "test2007",
    "train2007",
    "train2012",
    "val2007",
    "val2012"
]
```

```python
import shutil

def loop():
    cache = set()
    for dir in dirs:
        full_path = os.path.join(data_input, "images", dir)
        for file_ in os.listdir(full_path):
            name, _ = file_.split(".")
            cache.add(name)
            if len(cache) == 1000:
                return cache

            # save to output
            img_path = os.path.join(full_path, file_)
            output_path = os.path.join(data_output, "images")
            shutil.copy(img_path,output_path)
cache = loop()
```

```python
def loop_labels(cache):
    for dir in dirs:
        full_path = os.path.join(data_input, "labels", dir)
        for file_ in os.listdir(full_path):
            name, _ = file_.split(".")
            if name in cache:
                print(file_)
                cache.remove(name)
                # save to output
                file_path = os.path.join(full_path, file_)
                output_path = os.path.join(data_output, "labels")
                shutil.copy(file_path,output_path)
                print(len(cache))


loop_labels(cache)
```
